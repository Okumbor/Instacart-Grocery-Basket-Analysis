```python
# creating a function to categorise the product prices

def price_label(row):
    
    if row['prices'] <= 5:
        return "Low range price"
    elif (row['prices'] >5) and (row['prices']<= 15):
        return "Medium range price"
    elif row['prices'] >15:
        return "High range price"
    else:
        return "Not enough data"
```

```python
df['price_range'] = df.apply(price_label, axis = 1)
```

```python
df['price_range'].value_counts()
```

```python
# counting the days of the week and sort it in ascending order

df_ord_prod['order_days_of_week'].sort_values(ascending = True).value_counts(dropna = False)
```

```python
# allocating the busiest day to 0 and the least busy to day 4

result = []

for value in df_ord_prod['order_days_of_week']:
    if value == 0:
        result.append('Busiest day')
    elif value == 4:
        result.append('Least busy day')
    else: 
        result.append('Regular busy day')
```

#### When too many users make Instacart orders at the same time, the app freezes. 
#### The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, 
#### they want periods of time labeled Most orders, Average orders, and Fewest orders. 
#### Create a new column containing these labels called busiest_period_of_day.

```python
df_ord_prod['order_hour_of_day'].sort_values(ascending = True).value_counts(dropna = False)
```

```python
df_ord_prod.loc[df_ord_prod['order_hour_of_day'].isin([10,11,14,15,13,12,16,9,17]), 'busiest_period_of_day'] = 'Most orders' 
```

```python
df_ord_prod.loc[df_ord_prod['order_hour_of_day'].isin([8,18,19,20,7,21,22,23,6,0,1]), 'busiest_period_of_day'] = 'Average orders'
```

```python
df_ord_prod.loc[df_ord_prod['order_hour_of_day'].isin([5,2,4,3]), 'busiest_period_of_day'] = 'Fewest orders'
```

#### Print the frequency for this new column.

```python
df_ord_prod['busiest_period_of_day'].value_counts(dropna = False).sort_values(ascending = False)
```
